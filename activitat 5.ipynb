```python
class song():
    def __init__(self,titol,autor,durada):
        self.titol=titol
        self.autor=autor
        self.durada=durada
    def info(self):
        return f"{self.titol}-{self.autor}-({self.durada} segons)"

class playlist:

    def __init__ ( self, song_list):
        self.song_list = song_list  
        song_list = []

    def afegir_canço ( self,song ):
        self.song_list.append(song)

    def durada_total ( self):
        total = 0
        for i in self.song_list:
            total = total+i.durada
        return f"{total} segons"

    def mostrar_playlist ( self):
        for n in self.song_list:
            print ( n.info())

    def canço_mes_llarga ( self):
        canço_mes_llarga = self.song_list[0]
        for x in self.song_list:
            if x.durada > canço_mes_llarga.durada:
                canço_mes_llarga = x
        return canço_mes_llarga.info()


song1=song("DJ got us falling in love", "Usher", 220 )

song2 = song ( "Scream and shout", "Britney spears", 283)

playlist1 = playlist ( [song1, song2])

song4 = song ( "Everybody", "Backstreet Boys", 225 )  

playlist1.afegir_canço ( song4)

song3= song("Kiss me thru the phone","Soujla Boy", 193)
playlist1.afegir_canço(song3)


print ( playlist1.durada_total())
print ( playlist1.canço_mes_llarga())
playlist1.mostrar_playlist()



```

```python
class personatge():
    def __init__(self,nom,vida,força,nivell):
        self.nom=nom
        self.vida=vida
        self.força=força
        self.nivell=nivell
    def atacar(self,enemic):
        enemic.vida=enemic.vida - self.força
    def info(self):
        return f"{self.nom}-{self.vida}-{self.força}-{self.nivell}"    
    def pujar_nivell(self):
        self.vida=self.vida+1
        self.força=self.força+5
        self.nivell=self.nivell+1

    def esta_viu(self):
        if self.vida>0:
            return True
        else:
            return False
    def __str__(self):
        return f"Nom:{self.nom}, Vida:{self.vida}\nForça:{self.força}, Nivell:{self.nivell}"
    
jugador=personatge("jugador1", 10,2,1)
enemic=personatge("enemic",8,3,1)
compt =0
while (jugador.vida or enemic.vida > 0 )and compt < 1000:
    compt+=1
    jugador.atacar(enemic)
    if enemic.esta_viu():
        enemic.atacar(jugador)
    elif jugador.esta_viu() == False:
        print("Has perdut")
        break
    else:
        print("Has guanyat!!!")
        jugador.pujar_nivell()
        break
jugador.info()
print(jugador)










```
